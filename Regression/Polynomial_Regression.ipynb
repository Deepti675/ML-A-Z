```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('/content/Position_Salaries.csv')
x=dataset.iloc[:,1:-1].values
y=dataset.iloc[:,-1].values
```

```python
print(x)
```

```python

# from sklearn.model_selection import train_test_split
# x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(x,y)
#y=b0+b1x1
```

```python
#y=b0+b1x1+b2x2^2+.........................
#polynomial regression model
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=3)
#new matrix feature with degree 2 like x1,x1^2
x_poly=poly_reg.fit_transform(x)
#now implement regression model with new transform feature
lin_reg2=LinearRegression()
lin_reg2.fit(x_poly,y)
```

**visualising the results**

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg.predict(x),color='blue')
plt.title('Truth linear regression')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg2.predict(x_poly),color='blue')
plt.title('Truth polynomial regression')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

Predicting new result with linear model

```python
lin_reg.predict([[6.5]])
```

Predict new result with polynomial model

```python
lin_reg2.predict(poly_reg.fit_transform([[6.5]]))
```
