```python
#k mean clustering
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('/content/Mall_Customers.csv')
x=dataset.iloc[:,[3,4]].values
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(x, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
bc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')
y_bc = bc.fit_predict(x)
```

```python
plt.scatter(x[y_bc==0,0],x[y_bc==0,1],s=100,c='red',label='cluster 1')
plt.scatter(x[y_bc==1,0],x[y_bc==1,1],s=100,c='blue',label='cluster 2')
plt.scatter(x[y_bc==2,0],x[y_bc==2,1],s=100,c='green',label='cluster 3')
plt.xlabel('Cluster of customers')
plt.xlabel('Annual Income')
plt.ylabel('spending money')
```
